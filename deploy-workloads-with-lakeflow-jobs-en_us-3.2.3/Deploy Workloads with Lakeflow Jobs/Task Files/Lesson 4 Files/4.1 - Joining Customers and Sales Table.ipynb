```python
%sql
CREATE OR REPLACE TABLE customers_sales_silver AS
SELECT
  c.customer_id,
  c.customer_name,
  c.loyalty_segment,
  c.units_purchased,
  s.product_name,
  s.order_date,
  s.total_price
FROM
  customers_bronze c
JOIN
  sales_bronze s
ON
  c.customer_id = s.customer_id;
```

```python
df = spark.sql("""
    SELECT * FROM customers_sales_silver
""")

# Check for duplicate records
duplicate_exists = df.count() > df.dropDuplicates().count()

# Set boolean flag in task values
dbutils.jobs.taskValues.set(key="has_duplicates", value=duplicate_exists)
```
