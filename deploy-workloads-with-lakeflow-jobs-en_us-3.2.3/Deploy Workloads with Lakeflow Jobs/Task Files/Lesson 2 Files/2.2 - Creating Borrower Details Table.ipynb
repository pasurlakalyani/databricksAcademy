```python
%sql
-- Create the Borrowers Table
CREATE TABLE IF NOT EXISTS borrower_details_silver (
    id INT PRIMARY KEY,
    age INT,
    yoe INT,
    income_in_usd INT,
    zip_code INT,
    family_size INT,
    education STRING
);
```

```python
%sql
-- Deleting Records before Insertion
DELETE FROM borrower_details_silver
```

```python
%sql
-- Transforming and Inserting records into our newly created table
INSERT INTO borrower_details_silver
SELECT 
    id,
    age,
    experience AS yoe,
    cast(income AS FLOAT) * 1000 AS income_in_usd,
    zip_code,
    family AS family_size,
    CASE 
        WHEN education = '1' THEN 'Undergraduate'
        WHEN education = '2' THEN 'Graduate'
        WHEN education = '3' THEN 'Postgraduate'
    END AS education
FROM bank_master_data_bronze;
```
